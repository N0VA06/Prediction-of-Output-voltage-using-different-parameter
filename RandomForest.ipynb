```python
X= dataset.drop(['U_tot (V)'],axis=1)
X
```

```python
scalar = MinMaxScaler()
X_scaled = scalar.fit_transform(X)
```

```python
Y = dataset.drop(['t (sec)','i (A)', 'P (W)', 'm_Air', 'm_H2', 'RH_Air','RH_H2', 'P_Air_inlet', 'P_H2_inlet', 'P_Air_outlet', 'P_H2_outlet','T_1', 'T_2', 'T_3', 'T_4', 'T_Air_inlet', 'T_H2_inlet',
       'T_Stack_outlet', 'T_Heater', 'm_Air_write', 'm_H2_write',
       'Heater_power', 'i_write'],axis=1)
Y
```

```python
Y = Y.values.reshape(-1,1)
Y
```

```python
y_scaled = scalar.fit_transform(Y)
y_scaled
```

```python
X_scaled = scalar.fit_transform(X)
X_scaled
```

```python
print(X_scaled.shape)
print(y_scaled.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size = 0.25,random_state=42)
X
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators=100, random_state=42)
reg.fit(X_train, y_train)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
rmse = np.sqrt(mse)
print('RMSE: ',rmse)
```
